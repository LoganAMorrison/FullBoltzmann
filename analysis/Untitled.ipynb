```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt("full_boltz_data.dat")
xs = data[:,0]
fs = data[:,1:]
qs = np.linspace(1e-6, 50.0, len(fs[-1]))
```

```python
plt.figure(dpi=150)
for i in range(0,len(xs),1000):
    integral = np.trapz(qs**2 * fs[i], x=qs)
    plt.plot(qs, qs**2 * fs[i] / integral, label=f'x={xs[i]}')
plt.xlim([0.0,25.0])
plt.legend()
```

```python
qs = np.linspace(1e-6, 50.0, len(fs[-1]))

mdm = 100.0
T0 = 2.3525e-13*(2.725/2.73);
rhoc = 8.0980e-47;
xB = 100.0
x0 = mdm/T0

nBs = []
xx=[]
for i in range(0,len(xs),500):
    xx.append(xs[i])
    intB = np.trapz(qs**2 * fs[i], x=qs)
    nBs.append(intB / (2.0 * np.pi**2) * (mdm / xB)**3*  mdm * (xB / x0)**3 / rhoc)
#nB * mdm * (xB / x0)**3 / rhoc
```

```python
plt.plot(xx,nBs)
plt.yscale('log')
plt.xscale('log')
```
